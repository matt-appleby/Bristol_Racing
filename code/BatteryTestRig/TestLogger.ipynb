```python
#hook up to venv racing

#needs pyserial not serial
#needs arduino ide closed

import numpy as np
import matplotlib.pyplot as plt
#import visa
#import pyvisa
from matplotlib import _pylab_helpers
#import tkinter
from tkinter import filedialog
#import threading
#import csv
import pandas as pd
import serial
import time
```

```python
#Sets up the dataframe and wiping old data
#data= pd.DataFrame(columns=['Time','Current','Voltage'])
data=pd.DataFrame();
data
```

```python
# run this loop unitl interupt 

#if it errors run lower code cell to close comm

serialcomm = serial.Serial("COM6", 9600, timeout=None)
String2=""
while True:
    String1 =  serialcomm.readline().decode('ascii')
    if String1 != String2:
        ListedString = list(String1.split(" "))
        if len(ListedString)>2:
            ListedString[2]=(ListedString[2])[:-3]
        print(ListedString)
        TimeInstanceData = pd.DataFrame(ListedString)
        TimeInstanceData = TimeInstanceData.transpose()
        data = pd.concat([data, pd.DataFrame.from_records(TimeInstanceData)])
    time.sleep(0.5)
    String2=String1
```

```python
# close serial comm and display
serialcomm.close()
data
```

```python
BatteryRef=input("What is the battery reference letter (A,B)")
Date=input("What is the date (12 08 09 2022) style in hour day month year")
data.to_csv("DischargeData/BatteryDischargeTest"+BatteryRef+Date+'.csv')
```
